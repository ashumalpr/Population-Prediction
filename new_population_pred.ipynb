```python
# we have to create 4 different data sets for 4 different model
#columns_for_ = ['Year', 'Population','Life Expectancy','Birth Rate','Infant Mortality Rate']

# Create a new DataFrame for 'Life Expectancy' model
df_life_expectancy = df[['Year','Life Expectancy']]

# Create a new DataFrame for 'Birth Rate'  model
df_birth_rate = df[['Year','Birth Rate']]

# Create a new DataFrame for 'Infant Mortality Rate' model
df_infant_mortality_rate = df[['Year','Infant Mortality Rate']]

# Create a new DataFrame for master model
df_m_model = df[['Year','Population','Life Expectancy','Birth Rate','Infant Mortality Rate']]
```

```python
print(df_life_expectancy.shape)
print(df_birth_rate.shape)
print(df_infant_mortality_rate.shape)
print(df_m_model.shape)
```

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

## Model for predicting 'Life Expectancy' overs the year

```python
x = sm.add_constant(df_life_expectancy['Year'])
y = df_life_expectancy['Life Expectancy']
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x , y, train_size = 0.8,test_size = 0.2,random_state = 100)
```

```python
# fitting the model 
life_expectancy_lm=sm.OLS(y_train,x_train).fit()
```

```python
print(life_expectancy_lm.params)
```

```python
print(life_expectancy_lm.summary())
```

```python
# predict values of test & train dataset
y_pred_test  = life_expectancy_lm.predict(x_test)
y_pred_train = life_expectancy_lm.predict(x_train)
```

```python
# measuring accracy of prediction
```

```python
# R square
print(np.abs(r2_score(y_test,y_pred_test)))
print(np.abs(r2_score(y_train,y_pred_train)))
```

```python
# RMSE
print(np.sqrt(mean_squared_error(y_train,y_pred_train)))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_train, y_pred_train, c='blue', label='Data points')
plt.plot([y_train.min(), y_train.max()], [y_train.min(), y_train.max()], 'k--', lw=3, label='Ideal fit')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Values")
plt.legend()
plt.show()
```

```python
comparison_l = pd.DataFrame({'actual': y_train , 'predicted': y_pred_train ,'error %': round(100*(y_train-y_pred_train)/y_train,2)})
print(comparison_l)
```

## Master Model for predicting 'Population' overs the year

```python
x = df_m_model[['Year','Life Expectancy','Birth Rate','Infant Mortality Rate']].values
y = df_m_model['Population'].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x , y, train_size = 0.8,test_size = 0.2,random_state = 100)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
#representing LInearRegression as lr
lr = LinearRegression()
#fit the model using lr.fit()
lr.fit(x_train, y_train)
```

```python
print(round(lr.intercept_,2))
print(lr.coef_)
```

```python
y_pred = lr.predict(x_test)
```

```python
# Create DataFrame
comparison = pd.DataFrame({'actual': y_test , 'predicted': y_pred ,'error %': 100*(y_test-y_pred)/y_test})

# Convert 'predicted' to integers
comparison['predicted'] = comparison['predicted'].round().astype(int)
comparison['error %'] = comparison['error %'].round(3)

print(comparison)
```

```python
def pred_life_expectancy(year) :                    # life_expectency_lm.params
    return round(0.494159 * year - 926.627457,2)
def pred_birth_rate(year) :                         # birth_rate_lm.params
    return round(-0.402001 * year+ 830.497016 ,3)
def pred_infant_mortality_rate(year) :              
    return round(-2.318530 * year + 4706.718052,3)  # infant_mortality_rate.params


def predict_all(year):
    life_expectancy = pred_life_expectancy(year) 
    birth_rate = pred_birth_rate(year)
    infant_mortality_rate = pred_infant_mortality_rate(year)
    
    return year,life_expectancy, birth_rate, infant_mortality_rate
```

```python
# Predict population from 2023 to 2033
for year in range(2023, 2036):
    # Get the predictions from the predict_all function and convert to a list
    features = list(predict_all(year))
    predicted_population = int(lr.predict([features[0:]])[0])
    print(f"{year}: {predicted_population:,}")
```
